```python
# Data Setup

import matplotlib.pyplot as plt
import numpy as np

ml=np.array([10813,10082,12492,15204,18199,14813,15664,17528,20997,24921,30733,35689,47101,62509,78678])

grav=np.array([1614,1872,1916,2039,2144,2512,2783,2773,3322,3723,4038,5377,6511,8961,8849])

years=np.array([2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])

ml=ml/1000
grav=grav/1000

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111, label="1")
ax2=fig.add_subplot(111, label="2", frame_on=False)

mline=ax.plot(years,ml, color='red',label='Search Term: Machine Learning',lw=3)
ax.set_xlabel("Publication Year",size=20)
ax.set_ylabel("Publication number [x1000]", color='red',size=20)
ax.tick_params(axis='y', colors="red")

gl=ax2.plot(years, grav,label='Search Term: Gravitational Waves',lw=3)
ax2.yaxis.tick_right()
ax2.set_ylabel('Publication number [x1000]', color="blue",size=20)       
ax2.yaxis.set_label_position('right') 
ax2.tick_params(axis='y', colors="blue")



lns = mline+gl
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0, fontsize=17)

plt.annotate('1st direct detection',(2015,4.038),xytext=(2010,5),arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),fontsize=16)


plt.savefig('wos.png',bbox_inches='tight')
plt.show()


```
